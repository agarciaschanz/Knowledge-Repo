```python
 # Import libraries and dependencies
import pandas as pd
from pathlib import Path
%matplotlib inline
```

```python
# Set file path
nflx_data = Path("nflx.csv")

# Read the CSV into a DataFrame and set the 'date' column as a datetime index
nflx_df = pd.read_csv(nflx_data, index_col="date", infer_datetime_format=True, parse_dates=True)
nflx_df.sort_index(inplace=True)

# Display a few records
nflx_df.head()
```

```python
# Plot DataFrame
nflx_df.plot()
```

```python
 # Calculate 20-day Simple Moving Average for NFLX Closing Prices
nflx_sma_20 = nflx_df.rolling(window=20).mean()
nflx_sma_20.head(50)
```

```python
# Calculate 50-day Simple Moving Average for NFLX Closing Prices
nflx_sma_50 = nflx_df.rolling(window=50).mean()
nflx_sma_50.head(50)
```

```python
# Calculate 100-day Simple Moving Average for NFLX Closing Prices
nflx_sma_100 = nflx_df.rolling(window=100).mean()
nflx_sma_100.head(50)
```

```python
# Set figure of the daily closing prices of NFLX
ax = nflx_df.plot(figsize=(25,10))

# Overlay SMA20, SMA50, and SMA100 on the same figure
nflx_sma_20.plot(ax=ax)
nflx_sma_50.plot(ax=ax)
nflx_sma_100.plot(ax=ax)

# Set the legend of the figure
ax.legend(["Original", "SMA20", "SMA50", "SMA100"]);
```

```python
# Set figure of the daily closing prices of NFLX
ax = nflx_df.plot(figsize=(25,10))

# Overlay STD20, STD50, and STD100 on the same figure
nflx_df.rolling(window=20).std().plot(ax=ax)
nflx_df.rolling(window=50).std().plot(ax=ax)
nflx_df.rolling(window=100).std().plot(ax=ax)

# Set the legend of the figure
ax.legend(["Original", "STD20", "STD50", "STD100"]);
```
